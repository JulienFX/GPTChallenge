```python
import numpy as np 
a = sorted(np.random.randint(0,100,size=10))
print(a)
import random
b = sorted([random.randint(0, 99) for _ in range(10)])
print(b)
communs = [x for x in a if x in b]

if communs:
    print("oui")
    print("Valeurs communes :", communs)
else:
    print("non")
```

```python
communs = list(set(a) & set(b))

if communs:
    print("oui")
    print("Valeurs communes :", communs)
else:
    print("non")
```

```python
class Randomi : 
    def __init__(self) : 
        self.a = sorted(np.random.randint(0,100,size=10))
        self.b = sorted([random.randint(0, 99) for _ in range(10)])
    def InOrNot(self) : 
        return any(x in self.a for x in self.b) # si au moins une valeur de a est dans b 
dam = Randomi()
dam.InOrNot()
```
