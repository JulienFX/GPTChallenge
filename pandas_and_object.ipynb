## Exercice 1 

```python
class Produit : 
    def __init__(self,name,price,quantity) : 
        # attributs d'instance 
        self.nom = name
        self.prix = price
        self.qte = quantity
    def valeur_stock(self) : 
        return self.prix * self.qte
    def __str__(self) : 
        return f"Produit : {self.nom}, Prix : {self.prix}€, Quantité : {self.qte}"
```

```python
produit = Produit("Clavier",20,5)
print(produit.valeur_stock())
print(produit) #pareil que print(__str__())
```

## Exercice 2 

```python
import pandas as pd 
class Inventaire : 
    def __init__(self) :
        self.df = pd.DataFrame(columns=['nom', 'prix', 'quantite'])
        
    def ajouter_produit(self,name,price,quantity) : 
        self.df.loc[len(self.df)] = [name, price, quantity]

    def valeur_totale(self) : 
        return (self.df['prix'] * self.df['quantite']).sum()
        
    def afficher_produits(self):
        return self.df
```

```python
i = Inventaire()
i.ajouter_produit("Ordinateur",550,10)
i.ajouter_produit("Souris",14,7)
print(i.afficher_produits())
print(i.valeur_totale())
```
