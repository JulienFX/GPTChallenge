## Exercice 1 : Objet -> Class

```python
class Produit : 
    def __init__(self,name,price,quantity) : 
        # attributs d'instance 
        self.nom = name
        self.prix = price
        self.qte = quantity
    def valeur_stock(self) : 
        return self.prix * self.qte
    def __str__(self) : 
        return f"Produit : {self.nom}, Prix : {self.prix}€, Quantité : {self.qte}"
```

```python
produit = Produit("Clavier",20,5)
print(produit.valeur_stock())
print(produit) #pareil que print(__str__())
```

## Exercice 2 : Objet & DataFrame

```python
import pandas as pd 
class Inventaire : 
    def __init__(self) :
        self.df = pd.DataFrame(columns=['nom', 'prix', 'quantite'])
        
    def ajouter_produit(self,name,price,quantity) : 
        self.df.loc[len(self.df)] = [name, price, quantity]

    def valeur_totale(self) : 
        return (self.df['prix'] * self.df['quantite']).sum()
        
    def afficher_produits(self):
        return self.df
    
    def produits_chers(self, prix_min) : 
        return self.df[self.df['prix']> prix_min]
    
    def trier_par_prix(self) : 
        return self.df.sort_values(by='prix', ascending=True)
```

```python
i = Inventaire()
i.ajouter_produit("Ordinateur",550,10)
i.ajouter_produit("Souris",14,7)
print(i.afficher_produits())
print(i.valeur_totale())
print(i.produits_chers(200))
print(i.trier_par_prix())
```

## Exercice 8 

```python
import pandas as pd
from datetime import datetime

ventes = pd.DataFrame({
    "commande_id": [1, 1, 2, 3, 3, 3, 4],
    "produit": ["A", "B", "A", "A", "C", "C", "B"],
    "quantite": [1, 2, 3, 1, 2, 2, 1],
    "prix_unitaire": [100, 50, 100, 100, 200, 200, 50],
    "date": [
        "2024-01-01", "2024-01-01",
        "2024-01-02",
        "2024-01-03", "2024-01-03", "2024-01-03",
        "2024-01-04"
    ]
})
produits = pd.DataFrame({
    "produit": ["A", "B", "C", "D"],
    "categorie": ["Tech", "Accessoire", "Tech", "Service"],
    "cout_unitaire": [60, 20, 120, 10]
})
```

```python
ventes.head()
```

```python
produits.head()
```

```python
ventes['date'] = pd.to_datetime(ventes['date'])
```

```python
ventes = ventes.drop_duplicates(subset=["commande_id", "produit", "quantite", "date"])
produits = produits.drop_duplicates()
```

```python
ventes.drop_duplicates(['produit','prix_unitaire']).filter(['produit','prix_unitaire']) # filter -> choix col affichage
```

```python
ventes['chiffre_affaires_ligne'] = ventes['prix_unitaire'] * ventes['quantite']
```

```python
ventes.head()
```

```python
df_merge = produits.merge(ventes, on='produit')
df_merge
```

```python
df_merge['marge'] = df_merge['chiffre_affaires_ligne'] - df_merge['cout_unitaire']
df_merge[df_merge['marge'] > 100] 
```

```python
marge_par_categorie = df_merge.groupby('categorie')['marge'].sum() #group by permet de grouper uniquement en vue de préparer une aggrégation donc rien ne se pase ici
marge_par_categorie
```

```python
a = df_merge.groupby('produit')['prix_unitaire'].sum().sort_values(ascending=False)[:2] # top 2 produits les plus rentables 
a
```
