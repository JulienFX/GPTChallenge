Numpy > List built-in library <br>
Numpy arrays are homogeneous, meaning all elements must be the same type allowing efficient computation <br>
Vectorized operations in NumPy can be 10 to 100 times faster than equivalent Python loops.

```python
import numpy as np
```

```python
a0 = np.zeros((3, 3))
a1  = np.ones((2, 6)) 
ar  = np.arange(0, 11, 2) # start end step

print(a0)
print(a1)
print(ar)
print(ar[-1])
```

```python
a = np.array([10, 20, 30, 40, 50, 60])
idx = np.arange(0,6,2)
print(a[idx])
cond = a> 30
print(a[cond])
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])


print(x + y)    # add
print(x - y)    # subtract
print(x * y)    # multiply
print(x / y)    # divide
```

## String manipulation 

```python
txt = "Hello World"[:-2]
print(txt) 
txt = "Hello World"[::-2] # prends un caractère sur 2 en partant de la fin 
print(txt)
txt = "Hello World"[::-1]
print(txt) 
import re 
txt = "Hello World"
txt = re.sub("[aeiou]", "", txt)
print(txt)
```
